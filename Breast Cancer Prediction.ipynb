```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import mutual_info_classif

# Ensure pandas displays all rows
pd.set_option('display.max_rows', None)

# Calculate mutual information
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']
mutual_info = mutual_info_classif(X, y)

# Create a DataFrame to rank features
mutual_info_df = pd.DataFrame({'Feature': X.columns, 'Mutual_Info': mutual_info})
mutual_info_df = mutual_info_df.sort_values(by='Mutual_Info', ascending=False)


# Plot the mutual information values
plt.figure(figsize=(12, 8))
sns.barplot(x='Mutual_Info', y='Feature', data=mutual_info_df, palette='viridis')
plt.title('Feature Importance based on Mutual Information', fontsize=16)
plt.xlabel('Mutual Information', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Apply Standard Scaling to the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert the scaled features back into a DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

```

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

class ModelTrainer:
    def __init__(self, X_train, X_test, y_train, y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test

    def train_random_forest(self):
        rf = RandomForestClassifier(
            n_estimators=200,
            max_features='sqrt',
            max_depth=None,
            min_samples_split=2,
            min_samples_leaf=1,
            bootstrap=True,
            random_state=42
        )
        rf.fit(self.X_train, self.y_train)
        y_pred = rf.predict(self.X_test)
        accuracy = accuracy_score(self.y_test, y_pred)
        report = classification_report(self.y_test, y_pred)
        return accuracy, report

    def train_xgboost(self):
        xgb_model = xgb.XGBClassifier(
            n_estimators=200,
            max_depth=6,
            learning_rate=0.1,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42
        )
        xgb_model.fit(self.X_train, self.y_train)
        y_pred_xgb = xgb_model.predict(self.X_test)
        accuracy_xgb = accuracy_score(self.y_test, y_pred_xgb)
        report_xgb = classification_report(self.y_test, y_pred_xgb)
        return accuracy_xgb, report_xgb

    def train_logistic_regression(self):
        log_reg = LogisticRegression(
            max_iter=1000,
            solver='liblinear',
            random_state=42
        )
        log_reg.fit(self.X_train, self.y_train)
        y_pred_log_reg = log_reg.predict(self.X_test)
        accuracy_log_reg = accuracy_score(self.y_test, y_pred_log_reg)
        report_log_reg = classification_report(self.y_test, y_pred_log_reg)
        return accuracy_log_reg, report_log_reg

    def train_svm(self):
        svm_model = SVC(C=1.0, kernel='rbf', random_state=42)
        svm_model.fit(self.X_train, self.y_train)
        y_pred_svm = svm_model.predict(self.X_test)
        accuracy_svm = accuracy_score(self.y_test, y_pred_svm)
        report_svm = classification_report(self.y_test, y_pred_svm)
        return accuracy_svm, report_svm

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

trainer = ModelTrainer(X_train, X_test, y_train, y_test)

# Train and evaluate RandomForest
accuracy_rf, report_rf = trainer.train_random_forest()
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
print("Random Forest Classification Report:")
print(report_rf)

# Train and evaluate XGBoost
accuracy_xgb, report_xgb = trainer.train_xgboost()
print(f"XGBoost Accuracy: {accuracy_xgb:.4f}")
print("XGBoost Classification Report:")
print(report_xgb)

# Train and evaluate Logistic Regression
accuracy_log_reg, report_log_reg = trainer.train_logistic_regression()
print(f"Logistic Regression Accuracy: {accuracy_log_reg:.4f}")
print("Logistic Regression Classification Report:")
print(report_log_reg)

# Train and evaluate SVM
accuracy_svm, report_svm = trainer.train_svm()
print(f"SVM Accuracy: {accuracy_svm:.4f}")
print("SVM Classification Report:")
print(report_svm)
```
